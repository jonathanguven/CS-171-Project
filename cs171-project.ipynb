```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better-looking plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
```

```python
df = pd.read_csv('california_housing.csv')
df = df.drop(columns=['brokered_by', 'status', 'city'])
df = df.dropna()
print(f"Data shape after cleaning: {df.shape}")

df.head()
```

```python
# Sample 100 rows for visualization
df_sample = df.sample(n=100, random_state=42)

# Create two simple plots side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Price distribution
axes[0].hist(df_sample['price'], bins=20, edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Price ($)')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Price Distribution')
axes[0].ticklabel_format(style='plain', axis='x')

# House size vs price
axes[1].scatter(df_sample['house_size'], df_sample['price'], alpha=0.6)
axes[1].set_xlabel('House Size (sq ft)')
axes[1].set_ylabel('Price ($)')
axes[1].set_title('House Size vs Price')
axes[1].ticklabel_format(style='plain', axis='y')

plt.tight_layout()
plt.show()
```
